```python
# Import the required modules
import numpy as np
import matplotlib.pyplot as plt
# This makes the plots appear inside the notebook
%matplotlib inline
```

```python
from scipy.integrate import odeint

    # define a function which calculate the derivative
def dy_dx(y,x):
    return x - y
xs = np.linspace(0,5,100)
y0 = 1.0
ys = odeint(dy_dx, y0, xs)
ys = np.array(ys).flatten()
```

```python
# plot the numerical solution
plt.rcParams.update({'font.size':14}) 
plt.xlabel("x")
plt.ylabel("y")
plt.plot(xs,ys)
```

```python
y_exact = xs - 1 + 2*np.exp(-xs)
y_difference = ys - y_exact
plt.plot(xs,ys,xs,y_exact,"+");
```

```python
y_diff = np.abs(y_exact - ys)
plt.semilogy(xs,y_diff)
plt.ylabel("Error")
plt.xlabel("x")
plt.title("Error in numerical integration")
# note the logarithmic scale on the y-axis
```

```python
def dU_dx(U,x):
    # Here U is a vector such that y=U[0] and z=U[1]. This function should return
    # [y] 
    return [U[1], -2*U[1]-2*U[0] + np.cos(2*x)]
U0 = [0,0]
xs = np.linspace(0,10,200)
Us = odeint(dU_dx,U0,xs)
ys = Us[:,0]
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title("Damped Harmonic Oscillator")
plt.plot(xs,ys);
```

```python
a,b,c,d = 1, 1, 1, 1
def dP_dt(P, t):
    return[P[0]*(a-b*P[1]), -P[1]*(c - d*P[0])]

ts = np.linspace(0,12,100)
P0 = [1.5, 1.0]
Ps = odeint(dP_dt, P0, ts)
prey = Ps[:,0]
predators = Ps[:,1]
```

```python
plt.plot(ts, prey, '+', label = 'Rabbits')
plt.plot(ts, predators, 'x', label = 'Foxes')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend();
```

```python
plt.plot(prey,predators,'-')
plt.xlabel('Rabbits')
plt.ylabel('Foxes')
plt.title('Rabbits vs Foxes')
```

```python
# plot more curves in this system
ic = np.linspace(1.0, 3.0,21)
for r in ic:
    P0 = [r, 1.0]
    Ps = odeint(dP_dt, P0, ts)
    plt.plot(Ps[:,0],Ps[:,1],'-')
    
plt.xlabel('Rabbits')
plt.ylabel('Foxes')
plt.title('R vs F')
```

```python
# plot more curves in this system
ic = np.linspace(1.0, 3.0,21)
for r in ic:
    P0 = [1.5, r]
    Ps = odeint(dP_dt, P0, ts)
    plt.plot(Ps[:,0],Ps[:,1],'-')
    
plt.xlabel('Rabbits')
plt.ylabel('Foxes')
plt.title('R vs F')
```
